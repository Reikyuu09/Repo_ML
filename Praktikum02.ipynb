```python
import matplotlib.pyplot as plt


# ngambil data dari dataframe dan menaruhnya di variabel bernama data_height
data_height = df['Height']

# membuat histogram
n, bins, patches, = plt.hist(data_height, bins=5, color='blue', edgecolor='black')

# menambahkan tabel

plt.title('Histogram Nilai')
plt.xlabel('Height')
plt.ylabel('Frekuensi')

# Tampilkan rentang frekuensi di sumbi x

bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, [f'{bins[i]:.0f}-{bins[i+1]:.0f}' for i in range(len(bins)-1)])

# Tampilka histogram
plt.show()
```
